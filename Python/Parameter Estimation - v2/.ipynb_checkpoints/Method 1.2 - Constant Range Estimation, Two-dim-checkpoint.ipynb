```python
import numpy as np
import time
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.layers import Dense, Lambda, Concatenate, Flatten, RNN, GRU, Conv1D
#from tensorflow.compat.v1.keras.layers import CuDNNGRU
import os
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
```

# The model class - Here, the polynomial functions are defined 
$$f(x, y) = [ax^{N}, bx^{N-1}, cx^{N-2}, ..., n] \cdot [y^{n}, y^{n-1}, y^{n-2}, ..., 1]$$
$$N_{\theta}= N^{2}+1$$

```python
class f_model:
    def __init__(self, theta_range, x_range, y_range, N_x, N_y):
        #Theta_range is the span of parameters a, b and c. 
        #x_range is the (x_start_x_stop) values that are allowed for x to start 
        #N_x is how many points in x_range that will be trained on.
        self.theta_range = theta_range 
        self.x_range = x_range 
        self.y_range = y_range
        self.N_x = N_x 
        self.N_y = N_y
        self.x_full = np.linspace(x_range[0], x_range[1], N_x)
        self.y_full = np.linspace(y_range[0], y_range[1], N_y)
        self.X, self.Y = np.meshgrid(self.x_full, self.y_full)
    
    def calc_output_full_x(self, order=2): 
        funcs = []
        theta_rand = np.random.uniform(self.theta_range[0], self.theta_range[1], size=(order+1, order+1))
        for i in range(order+1):
            for j in range(order+1): 
                funcs.append(self.X**i * self.Y**j*theta_rand[i, j])
        f_vals = np.sum(funcs, axis=0)
        theta = theta_rand.flatten()
        return f_vals, theta
        
class Scaler_class: 
    def __init__(self): 
        self.X_scaler = Normalizer()
        self.y_scaler = Normalizer()
        
    def fit_transform(self, X_NN, y_NN): #Data shape = (N_sim, N_steps, N_var)
        y_data = self.y_scaler.fit_transform(y_NN)
        X_NN_temp = X_NN.reshape(-1, X_NN.shape[-1]**2)
        X_NN_temp = self.X_scaler.fit_transform(X_NN_temp)
        X_NN_temp = X_NN_temp.reshape(X_NN.shape)
        return X_NN_temp, y_data
    
    def transform(self, X_NN, y_NN): 
        y_data = self.y_scaler.transform(y_NN)
        X_NN_temp = X_NN.reshape(-1, X_NN.shape[-1]**2)
        X_NN_temp = self.X_scaler.transform(X_NN_temp)
        X_NN_temp = X_NN_temp.reshape(X_NN.shape)
        return X_NN_temp, y_data
```

```python
theta_range = (-5, 5)
x_range = (-3, 3)
y_range = (-3, 3)
N_x = 10 
N_y = 10
N_sim = 100000
mod = f_model(theta_range, x_range, y_range, N_x, N_y)

N_h1 = 128
N_h2 = 200
histories = []

for i in range(1, 11): 
    N_order = i
    X_NN = []
    y_NN = [] 
    for i in range(N_sim): 
        X, y = mod.calc_output_full_x(N_order) #Returns X as shape (N_vars, N_steps)
        X_NN.append(X) 
        y_NN.append(y)

    X_NN = np.array(X_NN) 
    y_NN = np.array(y_NN)

    Scaler = Scaler_class()
    X_sc, y_sc = Scaler.fit_transform(X_NN, y_NN)
    X_train, X_test, y_train, y_test  = train_test_split(X_sc, y_sc, test_size=0.2)

    #Making the NN model, Functional API model 
    inputs = keras.Input(shape=(N_x, N_y,), name="Variables")
    NN = Conv1D(N_h1, kernel_size=8, strides=1, activation='relu')(inputs)
    #NN = Dense(N_h1, activation='relu', name="NN1")(inputs)
    NN = Flatten()(NN)
    NN = Dense(N_h2, activation='relu', name="NN2")(NN)
    output = Dense((N_order+1)**2, name="Out1")(NN)

    model = keras.Model(inputs=inputs, outputs=output, name="Param_est_NN")
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
                  loss='mse', metrics=['accuracy'])

    history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), batch_size=32, 
                       callbacks=[tf.keras.callbacks.EarlyStopping(patience=5)])
    histories.append(history.history) 
```

```python
fig, axes = plt.subplots(4, 1)
for i, history in enumerate(histories):
    for el, ax in zip(history, axes): 
        ax.plot(history[el], label=f"{i+1}\'th order")
        ax.set_title(el)
        ax.legend()
        ax.grid()

fig.set_figheight(26)
fig.set_figwidth(12)
#fig.tight_layout(h_pad=-0.2)
plt.show()
```

```python
# Save a dictionary into a pickle file.
import pickle
 
pickle.dump( histories, open( "history_10_order_training_2_dim_v1.p", "wb" ) )
```

```python
accuracies = [] 
val_accuracies = []
for hist in histories: 
    accuracies.append(hist['accuracy'][-1])
    val_accuracies.append(hist['val_accuracy'][-1])
    
n = np.arange(1, len(accuracies)+1, 1)

plt.plot(n, accuracies, '*', label="Accuracies") 
plt.plot(n, val_accuracies, label="Validation Accuracies") 
plt.xticks(n)

plt.legend()
plt.grid()
plt.xlabel("N\'th order polynomial")
plt.ylabel("Accuracy")
plt.show()
```
