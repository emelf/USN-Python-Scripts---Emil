```python
import numpy as np
import time
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.layers import Dense, Lambda, Concatenate, Flatten, RNN, GRU
#from tensorflow.compat.v1.keras.layers import CuDNNGRU
import os
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
```

# The model class - Here, the polynomial functions are defined 
$$f(x) = ax^{n} + bx^{n-1} + cx^{n-2} + ... + N$$


```python
class f_model:
    def __init__(self, theta_range, x_range, N_x, dx):
        #Theta_range is the span of parameters a, b and c. 
        #x_range is the (x_start_x_stop) values that are allowed for x to start 
        #N_x is how many points in x_range that will be trained on.
        self.theta_range = theta_range 
        self.x_range = x_range 
        self.N_x = N_x 
        self.dx = dx
        self.x_full = np.linspace(x_range[0], x_range[1], N_x)
    
    def calc_output_limit_x(self, order=2): 
        #Randomly forwards and backwards
        k = np.random.choice([-1, 1])
        x_start = np.random.uniform(self.x_range[0], self.x_range[1]) 
        x = np.array([x_start + k*i*self.dx for i in range(self.N_x)])
        f, y = self.nth_order(x, order)
        return np.array([x, f]).T, y
    
    def nth_order(self, x, N): 
        #N is the highest polynomial order
        theta = self.give_rand_theta(N)
        polys = np.linspace(0, N, N+1)[::-1]
        f = np.array([theta[n]*x**polys[n] for n in range(N+1)]) #Calcs each order separately for all x's
        return f.sum(axis=0), theta #Sums the appropriate f(x) values to get one f(x) for all x's
        
    def give_rand_theta(self, N): 
        theta = []
        for i in range(N+1): 
            theta.append(np.random.uniform(self.theta_range[0], self.theta_range[1]))
        return theta
    
class Scaler_class: 
    def __init__(self): 
        self.X_scaler = Normalizer()
        self.y_scaler = Normalizer()
        
    def fit(self, X_NN, y_NN): #Data shape = (N_sim, N_steps, N_var)
        self.y_scaler = self.y_scaler.fit(y_NN)
        X_NN_temp = X_NN.reshape(-1, X_NN.shape[-1])
        self.x_scaler = self.X_scaler.fit(X_NN_temp)
        X_NN_temp = self.X_scaler.transform(X_NN_temp).reshape(X_NN.shape)
        return X_NN_temp, self.y_scaler.transform(y_NN)
    
    def transform(self, X_NN, y_NN): 
        X_NN_temp = X_NN.reshape(-1, X_NN.shape[-1])
        X_NN_temp = self.X_scaler.transform(X_NN_temp).reshape(X_NN.shape)
            
        return X_NN_temp, self.y_scaler.transform(y_NN)
```

# Generate training and testing data to the NNs 
Note that in this first instance, linear scaling is applied. 
$$f(x) = ax^{2} + bx + c$$

```python
theta_range = (-5, 5)
x_range = (-3, 3)
dx = 0.05 
N_x = 10 
N_sim = 100000
mod = f_model(theta_range, x_range, N_x, dx)

N_h1 = 100
N_h2 = 100
histories = []

for i in range(1, 5): 
    N_order = i
    X_NN = []
    y_NN = [] 
    for i in range(N_sim): 
        X, y = mod.calc_output_limit_x(N_order) #Returns X as shape (N_vars, N_steps)
        X_NN.append(X) 
        y_NN.append(y)

    X_NN = np.array(X_NN) 
    y_NN = np.array(y_NN)

    Scaler = Scaler_class()
    X_sc, y_sc = Scaler.fit(X_NN, y_NN)
    X_train, X_test, y_train, y_test  = train_test_split(X_sc, y_sc, test_size=0.2)

    #Making the NN model, Functional API model 
    inputs = keras.Input(shape=(N_x, X_NN.shape[-1], ), name="Variables")
    NN = GRU(N_h1, activation='relu', name="NN1")(inputs)
    #NN = Flatten()(NN)
    NN = Dense(N_h2, activation='relu', name="NN2")(NN)
    output = Dense(N_order+1, name="Out1")(NN)

    model = keras.Model(inputs=inputs, outputs=output, name="Param_est_NN")
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=5e-4),
                  loss='mse', metrics=['accuracy'])

    history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test), batch_size=64, 
                       callbacks=[tf.keras.callbacks.EarlyStopping(patience=5)])
    histories.append(history.history) 
```

```python
fig, axes = plt.subplots(4, 1)
for i, history in enumerate(histories):
    for el, ax in zip(history, axes): 
        ax.plot(history[el], label=f"{i+1}\'th order")
        ax.set_title(el)
        ax.legend()
        ax.grid()

fig.set_figheight(26)
fig.set_figwidth(12)
#fig.tight_layout(h_pad=-0.2)
plt.show()
```

```python
theta_range = (0, 10)
x_range = (0, 6)
dx = 0.05 
N_x = 10 
N_sim = 100000
x = np.linspace(x_range[0], x_range[1], N_x)

def f_exp(): 
    th = [np.random.uniform(theta_range[0], theta_range[1]) for i in range(4)]
    #Expression a*exp(b*x + c) + d
    f = th[0]*np.exp(th[1]*x + th[2]) + th[3]
    return f, y

def f_log(): 
    th = [np.random.uniform(theta_range[0], theta_range[1]) for i in range(4)]
    #Expression a*log(b*x + c+1) + d
    f = th[0]*np.log(th[1]*x + th[2]+1) + th[3]
    return f, y

def f_power_x(): 
    th = [np.random.uniform(theta_range[0], theta_range[1]) for i in range(4)]
    #Expression a*b**(x + c) + d
    f = th[0]*th[1]**(x + th[2]) + th[3]
    return f, y

def f_sin(): 
    th = [np.random.uniform(theta_range[0], theta_range[1]) for i in range(4)]
    #Expression a*sin(b*x + c) + d
    f = th[0]*np.sin(th[1]*x + th[2]) + th[3]
    return f, y

N_h1 = 400
N_h2 = 50
histories = []

funcs = [f_exp, f_log, f_power_x, f_sin]
for f in funcs:
    X_NN = []
    y_NN = [] 
    for i in range(N_sim): 
        X, y = f()
        X_NN.append(X) 
        y_NN.append(y)

    X_NN = np.array(X_NN) 
    y_NN = np.array(y_NN)

#     Scaler = Scaler_class()
#     X_sc, y_sc = Scaler.fit(X_NN, y_NN)
    X_sc = normalize(X_NN)
    y_sc = normalize(y_NN)
    X_train, X_test, y_train, y_test  = train_test_split(X_sc, y_sc, test_size=0.2)

    #Making the NN model, Functional API model 
    inputs = keras.Input(shape=(N_x,), name="Variables")
    NN = Dense(N_h1, activation='relu', name="NN1")(inputs)
    NN = Dense(N_h2, activation='relu', name="NN2")(NN)
    output = Dense(len(y_NN[0]), name="Out1")(NN)

    model = keras.Model(inputs=inputs, outputs=output, name="Param_est_NN")
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
                  loss='mse', metrics=['accuracy'])

    history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), batch_size=32, 
                       callbacks=[tf.keras.callbacks.EarlyStopping(patience=5)])
    histories.append(history.history) 
```
